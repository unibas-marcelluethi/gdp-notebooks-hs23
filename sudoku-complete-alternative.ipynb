### Das Sudoku-Spiel

Ein Sudoku Spiel besteht aus einem $9 \times 9$ Feld, welches in Quadrate der Grösse $3 \times 3$ unterteilt ist. In einigen Zellen sind Zahlen, andere sind Leer. Das Ziel des Spiels ist es, die Zellen so auszufüllen, dass jede der Zahlen 1 bis 9 genau einmal pro Zeile, einmal pro Spalte und einmal pro Quadrat benutzt wird. 

![sudoku](images/sudoku-combined.png)


```java
class Sudoku {

    /** Representation of the puzzle as array of arrays.
      * Each cell contains a number from 0 to 9.
      * The 0 means the cell is empty and needs to be filled.
      */
    public static int[][] createBoard() {
        int[][] board = new int[][] {
          {5, 3, 0, 0, 7, 0, 0, 0, 0},
          {6, 0, 0, 1, 9, 5, 0, 0, 0},
          {0, 9, 8, 0, 0, 0, 0, 6, 0},
          {8, 0, 0, 0, 6, 0, 0, 0, 3},
          {4, 0, 0, 8, 0, 3, 0, 0, 1},
          {7, 0, 0, 0, 2, 0, 0, 0, 6},
          {0, 6, 0, 0, 0, 0, 2, 8, 0},
          {0, 0, 0, 4, 1, 9, 0, 0, 5},
          {0, 0, 0, 0, 8, 0, 0, 7, 9}
        };
        return board;
    }


    /** Prints the board. */
    public static void printBoard(int[][] board) {
        for (int row = 0; row < board.length; row = row + 1) {
            for (int col = 0; col < board[0].length; col = col + 1) {
                System.out.print(board[row][col] + " ");
            }
            System.out.println();
        }
    }
   
    /** Checks if the number at a position is valid for the situation. */
    public static boolean isValid(int[][] board, int row, int col) {
        return
            isValidRow(board, row, col) &&
            isValidCol(board, row, col) &&
            isValidCell(board, row, col);
    }
    
    /** Checks if the row for the given position is valid. */
    public static boolean isValidRow(int[][] board, int row, int col) {
        int newValue = board[row][col];
        for(int i = 0; i < board[row].length; i = i + 1) {
            if (i != col && board[row][i] == newValue) {
                return false;
            }
        }
        return true;
    }
    
    /** Checks if the column for the given position is valid. */
    public static boolean isValidCol(int[][] board, int row, int col) {
        int newValue = board[row][col];
        for(int i = 0; i < board.length; i = i + 1) {
            if (i != row && board[i][col] == newValue) {
                return false;
            }
        }
        return true;
    }
    
    /** Checks if the 3x3 field containing the given position is valid. */
    public static boolean isValidCell(int[][] board, int row, int col) {
        int startRow = row - row % 3;
        int startCol = (col / 3) * 3;
        for (int ri = startRow; ri < startRow + 3; ri = ri + 1) {
            for (int ci = startCol; ci < startCol + 3; ci = ci + 1) {
                if ( ( (ri != row) || (ci != col) ) &&
                     board[row][col] == board[ri][ci] ) {
                    return false;
                }
            }
        }
        return true;
    }
    
    /** Searches for the next free field in the current situation.*/
    public static boolean findFreeCell(int[][] board) {
        for(int row = 0; row < board.length; row = row + 1) {
            for( int col = 0; col < board[row].length; col = col + 1) {
                if (board[row][col] == 0) { // check if field is empty
                    return tryAllNumbers(board, row, col); // try all numbers for empty field
                }
            }
        }
        
        return true; // no empty field found, we have a solution
    }
    
    /** Try all possible numbers from 1 to 9 at the given location. */
    public static boolean tryAllNumbers(int[][] board, int row, int col) {
        for (int i = 1; i < 10; i = i + 1) {
            board[row][col] = i; // test if the current number 'i' is valid for the situation
            if ( isValid(board, row, col) ) {
                if (solve(board)) { // try to solve the sudoku with the added number
                    return true; // yeah - we have found a solution
                } else {
                    // with i no solution was found for the remaining fields,
                    // test next i in next iteration
                }   
            } else {
                // i was not valid, test next i in next iteration
            }
        }
        
        board[row][col] = 0; // if no solution was found, set field as empty
        return false; // report that no solution was found
    }
    
    
    // solve the puzzle with backtracking
    static boolean solve(int[][] board) {
        return findFreeCell(board);
    }

    
    
    public static void main(String[] args) {
        int[][] board = createBoard();        
        
        solve(board);
        printBoard(board);
            
    }

    
}

Sudoku.main(new String[0]);
```
